```python
# wget https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt
with open('./data/input.txt', 'r', encoding='utf8') as f:
  text = f.read()
```

```python
print(text[:100])
```

```python
import tiktoken
enc = tiktoken.get_encoding('gpt2')
tokens = enc.encode(text)
print(tokens[:24])
```

```python
# we're going to arrange the tokens into batches with T long seuqences of tokens with B independent sequences
import torch
buf = torch.tensor(tokens[:24])
x = buf.view(4,6)
print(x)

# the first 6 tokens in first row, next 6 in next row, etc as independent rows
# label for target to calculate loss fn
# for the END token, we don't have the next correct token (so not enough information here)
```

```python
# create label that has the same size as x but has all the predictions
# stored as y
buf = torch.tensor(tokens[:24+1])
x = buf[:-1].view(4,6)
y = buf[1:].view(4,6)
print(x)
print(y)
```

```python
import math
# standard deviation grows inside the residual stream
x = torch.zeros(768)
n = 100 # e.g. 100 layers
for i in range(n):
    x += torch.randn(768) * n**-0.5

# these are roughly equivalent without scaling factor
print(x.std())
print(math.sqrt(n))
```
